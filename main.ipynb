```python
import pandas as pd

dataset = pd.read_csv("dataset.csv")

dataset.head()
```

```python
datasetdc = {}



for classname in dataset["Sports"].unique():
    datasetdc[classname] = dataset[dataset["Sports"] == classname][["Height","Weight"]].values.tolist()

print(datasetdc)
```

```python
from matplotlib import pyplot as plt



for classname in datasetdc:
    if classname == "Cricket":
        color = 'r'
    else:
        color = 'g'
    for sample in datasetdc[classname]:
        plt.scatter(sample[0],sample[1],color= color)
        plt.show
```

```python
import numpy as np
from collections import Counter

def KNN(newsample,k):
    distances = []
    for classname in datasetdc:
        for sample in datasetdc[classname]:
            distance = np.linalg.norm(np.array(newsample) - np.array(sample))
            distances.append([distance,classname])
    distances = sorted(distances)[:k]
    votes = []
    for dis in distances:
        votes.append(dis[1])
    voteresult =Counter(votes).most_common()[0][0]
    print(voteresult)
KNN([173,75],3)
```
